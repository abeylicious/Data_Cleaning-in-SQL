```python
# importing all the libraries
from bs4 import BeautifulSoup
import requests
import time
import datetime

import smtplib
```

```python
#connect to the webpage
url ='https://www.jumia.com.ng/danami-women-black-red-contrast-off-white-crop-hoodie-64832122.html'
headers ={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36","Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
```

```python
# getting the content of the webpage
webpage = requests.get(url, headers=headers)
webpage
```

```python
# getting the content of the webpage
webpage = requests.get(url, headers=headers)
page_contents1 = BeautifulSoup (webpage.content,'html.parser')
page_contents1
```

```python
# for better formatting
page_contents2 = BeautifulSoup(page_contents1.prettify())
page_contents2
```

```python
# to get the title, id= productTitle
title = page_contents2.find('h1').get_text().strip()
title
```

```python
# scrapping off the price with the class keyword
price2 = page_contents2.find(class_='-b -ubpt -tal -fs24 -prxs').get_text().strip()
price2
```

```python
# scrapping off the rating with the class keyword
ratings = page_contents2.find(class_='stars _m _al').get_text().strip()
ratings
```

```python
# stripping off the currency sign
price = price2[1:].strip()
# extracting the first three subset of the rating
rating = ratings[0:3]
print(title)
print(price)
print(ratings) 
# checking the data type
type(title)
type(price)
type(ratings)
```

```python
# adding a timestamp
import datetime
today = datetime.date.today() 
print(today)
```

```python
# automatinf an inserti oninto a csv, we create a csv file with these parameters
import csv
# specifying the column title/heads and the data
header = ['tltle','price','rating','date']
data = [title, price,rating,today]
type(data)
```

```python
#creating the csv file
with open ('jumaiwebscrapping_data2.csv','w',newline ='',encoding = 'UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```

```python
# enabling data append to this file
with open ('jumaiwebscrapping_data2.csv','a+', newline ='',encoding = 'UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(data) 
```

```python
# defining a function to automate the process and run it once a day
# to send an email (just for fun) when a price hits below or equal a certain level, an if statement was added with smtplb library  

def check_jumiaprice_data():
    url =' https://www.jumia.com.ng/danami-women-black-red-contrast-off-white-crop-hoodie-64832122.html'
    headers ={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36","Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    webpage = requests.get(url, headers=headers)
    page_contents1 = BeautifulSoup (webpage.content,'html.parser')
    page_contents2 = BeautifulSoup(page_contents1.prettify())
    title = page_contents2.find('h1').get_text().strip()
    price2 = page_contents2.find(class_='-b -ubpt -tal -fs24 -prxs').get_text().strip()
    ratings = page_contents2.find(class_='stars _m _al').get_text().strip()
    price = price2[1:].strip()
    rating = ratings[0:3]
    today = datetime.date.today() 
    import csv
    header = ['tltle','price','rating','date']
    data = [title, price,rating,today]
    with open ('jumaiwebscrapping_data2.csv','a+', newline ='',encoding = 'UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
    if (price <= '5000'):
        send_mail()
 
    
```
